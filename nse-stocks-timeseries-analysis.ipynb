### Task 1
#### 1. Create 4,16,....,52 week moving average(closing price) for each stock and index. This should happen through a function.

```python
# period is in weeks        
def calculate_weekly_moving_average(stock, period, dataframe, feature):
    weekly_scaled_df = dataframe.asfreq(f'W', method='pad').resample('W').mean()
    weekly_moving_avg_df = weekly_scaled_df.rolling(window=period).mean()
    plt.figure(figsize=(22,10))
    
    plt.plot(weekly_scaled_df[feature], label='Actual Weekly Mean')
    plt.plot(weekly_moving_avg_df[feature], label=f'{period} Week Average Mean')
    plt.xlabel("Date")
    plt.ylabel(f"Mean for '{feature}'")
    plt.title(f'{stock} Actual Mean & Periodic Moving Average')
    plt.legend()
    plt.show()
```

```python
calculate_moving_average('INFY', 4, infy_df, 'Close')
calculate_moving_average('TCS', 4, tcs_df, 'Close')
calculate_moving_average('NIFTY IT', 4, nifty_it_df, 'Close')
```

### 3. Create the following dummy time series:
#### 3.1 Volume shocks - If volume traded is 10% higher/lower than previous day - make a 0/1 boolean time series for shock, 0/1 dummy-coded time series for direction of shock.

```python
def volume_shocks(x):
    if x > 10:
        return 1
    else: 
        return 0


infy_df['volume_shocks'] = (infy_df.Volume.pct_change()*100).apply(lambda x: volume_shocks(x))
tcs_df['volume_shocks'] = (tcs_df.Volume.pct_change()*100).apply(lambda x: volume_shocks(x))
nifty_it_df['volume_shocks'] = (nifty_it_df.Volume.pct_change()*100).apply(lambda x: volume_shocks(x))
```

#### 3.2 Price shocks - If closing price at T vs T+1 has a difference > 2%, then 0/1 boolean time series for shock, 0/1 dummy-coded time series for direction of shock.

```python
# Price shocks
def price_shocks(x):
    if x > 2:
        return 1
    else: 
        return 0


infy_df['price_shocks'] = (infy_df.Close.pct_change()*100).apply(lambda x: price_shocks(x))
tcs_df['price_shocks'] = (tcs_df.Close.pct_change()*100).apply(lambda x: price_shocks(x))
nifty_it_df['price_shocks'] = (nifty_it_df.Close.pct_change()*100).apply(lambda x: price_shocks(x))
```

#### 3.3 Pricing black swan - If closing price at T vs T+1 has a difference > 2%, then 0/1 boolean time series for shock, 0/1 dummy-coded time series for direction of shock.

```python
# Price black swan
def pricing_black_swan(x):
    if x > 2:
        return 1
    else: 
        return 0


infy_df['pricing_black_swan'] = (infy_df.Close.pct_change()*100).apply(lambda x: pricing_black_swan(x))
tcs_df['pricing_black_swan'] = (tcs_df.Close.pct_change()*100).apply(lambda x: pricing_black_swan(x))
nifty_it_df['pricing_black_swan'] = (nifty_it_df.Close.pct_change()*100).apply(lambda x: pricing_black_swan(x))
```

#### 3.4 Pricing shock without volume shock - based on points a & b - Make a 0/1 dummy time series.


```python
infy_df['price_with_volume_shock'] = infy_df['volume_shocks'] & infy_df['price_shocks']
tcs_df['price_with_volume_shock'] = tcs_df['volume_shocks'] & tcs_df['price_shocks']
nifty_it_df['price_with_volume_shock'] = nifty_it_df['volume_shocks'] & nifty_it_df['price_shocks']
```
